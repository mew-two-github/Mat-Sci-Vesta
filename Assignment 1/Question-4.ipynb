```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sin, radians, sqrt
from sklearn.preprocessing import StandardScaler as SS
```

```python
X0_data = pd.read_excel('Cs2AgInCl6.xlsx',sheet_name = 'part b)')
X0_data = X0_data.loc[(X0_data['2θ']<=60) & (X0_data['2θ']>=20)]
X0_data.head()
```

```python
X0_data.info()
```

```python
#from sklearn.preprocessing import StandardScaler as SS
#scaler = SS()
```

```python
#X0_data['I_normalised'] = scaler.fit_transform(np.asarray(X0_data['I']).reshape(-1, 1))
X0_data['I_normalised'] = (X0_data['I']-min(X0_data['I']))/max(X0_data['I'])
X0_data.to_excel('normalised_x0.xlsx')
```

```python
plt.figure(figsize=(10,10))
plt.xlabel('2θ'); plt.ylabel('Intensity')
plt.plot(X0_data['2θ'],X0_data['I'])
```

```python
plt.figure(figsize=(10,10))
plt.xlabel('2θ'); plt.ylabel('Normalised Intensity'); plt.title('Experimental XRD')
plt.plot(X0_data['2θ'],X0_data['I_normalised'])
```

```python
X0_vesta = pd.read_excel('Cs2AgInCl6.xlsx',sheet_name = 'xrd_data')
X0_vesta = X0_vesta.loc[(X0_vesta['2θ']<=60) & (X0_vesta['2θ']>=20)]
#X0_vesta['I_normalised'] = scaler.fit_transform(np.asarray(X0_vesta['I']).reshape(-1, 1))
X0_vesta['I_normalised'] = X0_vesta['I']/max(X0_vesta['I'])
plt.figure(figsize=(10,10))
plt.xlabel('2θ'); plt.ylabel('Normalised Intensity'); plt.title('XRD from Vesta')
plt.plot(X0_vesta['2θ'],X0_vesta['I_normalised'])
```

```python
plt.figure(figsize=(10,10))
plt.xlabel('2θ'); plt.ylabel('Normalised Intensity'); plt.title('XRD from Vesta')
plt.plot(X0_data['2θ'],X0_data['I_normalised'],X0_vesta['2θ'],X0_vesta['I_normalised'])
plt.legend(["Experimental","Vesta"],loc ="upper right")
```

<h3>Lattice Parameters</h3>

```python
wvlength = 1.5406
theta = X0_data[X0_data.I == X0_data.I.max()]['2θ']
d = wvlength/(2*sin(radians(theta/2)))
a = d*sqrt(8)
```

```python
X0_data[X0_data.I == X0_data.I.max()]
```

```python
print('Distance between the planes: {} Angstrom \nLattice Parameter: {} Angstrom'.format(d,a))
```

<b>Conclusion:</b> 
<ol>
    <li>As seen from the peaks, and the data used in Excel in part a), the experimental data and the Vesta generated data  are <b>nearly identical</b> but not the same. </li>
    <li>Lattice Parameter from Vesta: <b>a = 10.48059 Angstrom</b> </li>
    <li>Lattice Parameter from Experimental data: <b>a = 10.49638 Angstrom</b> </li>

```python
lattice_params = []
wvlength = 1.5406
X0_data = pd.read_excel('XRD.xlsx',sheet_name = 'x=0')
X0_data = X0_data.loc[(X0_data['2θ']<=60) & (X0_data['2θ']>=20)]
theta = X0_data[X0_data.I == X0_data.I.max()]['2θ']
d = wvlength/(2*sin(radians(theta/2)))
a = d*sqrt(8)
lattice_params.append(a)
print(theta)
print('Distance between the planes: {} Angstrom \nLattice Parameter: {} Angstrom'.format(d,a))
```
